# Muestre algunos ejemplos de cada una de las modalidades que tiene

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
import os


m0_path =  "./archive/PolyMNIST/MMNIST/train/m0/"


m0_files = os.listdir(m0_path)[:5]  


fig, axes = plt.subplots(1, 5, figsize=(15, 5))

for j, file in enumerate(m0_files):
    image_path = os.path.join(m0_path, file)
    image = imread(image_path)
    axes[j].imshow(image)
    axes[j].axis('off')  
    axes[j].set_title(f'm0 - {file}')  


plt.tight_layout()
plt.show()
```
