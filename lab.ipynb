```python
import os
import numpy as np
from skimage.io import imread
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout


datos = [
    "./archive/PolyMNIST/MMNIST/train/m0/", "./archive/PolyMNIST/MMNIST/test/m0/", 
    "./archive/PolyMNIST/MMNIST/train/m1/", "./archive/PolyMNIST/MMNIST/train/m1/", 
    "./archive/PolyMNIST/MMNIST/train/m2/", "./archive/PolyMNIST/MMNIST/test/m2/", 
    "./archive/PolyMNIST/MMNIST/train/m3/", "./archive/PolyMNIST/MMNIST/test/m3/", 
    "./archive/PolyMNIST/MMNIST/train/m4/", "./archive/PolyMNIST/MMNIST/test/m4/"
]


imagenes = []
etiquetas = []


for dato in datos:
    archivos = os.listdir(dato)
    for archivo in archivos:
        ruta = os.path.join(dato, archivo)
        if os.path.exists(ruta):
            img = imread(ruta)
            imagenes.append(img)
            etiqueta = int(archivo.split('.')[1])  
            etiquetas.append(etiqueta)


imagenes = np.array(imagenes)
etiquetas = np.array(etiquetas)


imagenes = imagenes.astype('float32') / 255.0
etiquetas = to_categorical(etiquetas, 10)
x_train, x_test, y_train, y_test = train_test_split(imagenes, etiquetas, test_size=0.2, random_state=42)


modelo_simple = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])


modelo_simple.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


modelo_avanzado = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])


modelo_avanzado.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


historial_simple = modelo_simple.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=128)

historial_avanzado = modelo_avanzado.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=128)

# Evaluación del modelo simple
pérdida_simple, precisión_simple = modelo_simple.evaluate(x_test, y_test)
print(f'Modelo Simple - Precisión: {precisión_simple}')

# Evaluación del modelo avanzado
pérdida_avanzada, precisión_avanzada = modelo_avanzado.evaluate(x_test, y_test)
print(f'Modelo Avanzado - Precisión: {precisión_avanzada}')

if precisión_simple > precisión_avanzada:
    print("El Modelo Simple es mejor.")
else:
    print("El Modelo Avanzado es mejor.")
```

```python
import os
import numpy as np
from skimage.io import imread
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Cargar las imágenes y etiquetas desde las carpetas
datos = [
    "./archive/PolyMNIST/MMNIST/train/m0/", "./archive/PolyMNIST/MMNIST/test/m0/", 
    "./archive/PolyMNIST/MMNIST/train/m1/", "./archive/PolyMNIST/MMNIST/train/m1/", 
    "./archive/PolyMNIST/MMNIST/train/m2/", "./archive/PolyMNIST/MMNIST/test/m2/", 
    "./archive/PolyMNIST/MMNIST/train/m3/", "./archive/PolyMNIST/MMNIST/test/m3/", 
    "./archive/PolyMNIST/MMNIST/train/m4/", "./archive/PolyMNIST/MMNIST/test/m4/"
]

imagenes = []
etiquetas = []

# Iterar sobre las carpetas para cargar las imágenes y etiquetas
for dato in datos:
    archivos = os.listdir(dato)
    for archivo in archivos:
        ruta = os.path.join(dato, archivo)
        if os.path.exists(ruta):
            img = imread(ruta)
            imagenes.append(img.flatten())  # Aplanar las imágenes a 1D
            etiqueta = int(archivo.split('.')[1])  # Usar el segundo número del archivo como etiqueta
            etiquetas.append(etiqueta)

# Convertir listas a arrays de NumPy
imagenes = np.array(imagenes)
etiquetas = np.array(etiquetas)

# División del conjunto en entrenamiento y validación
x_train, x_val, y_train, y_val = train_test_split(imagenes, etiquetas, test_size=0.2, random_state=42)

# Definir el modelo Random Forest
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Entrenar el modelo
modelo_rf.fit(x_train, y_train)

# Predecir sobre el conjunto de validación
y_pred = modelo_rf.predict(x_val)

# Evaluar el modelo
precision = accuracy_score(y_val, y_pred)
print(f"Precisión de Random Forest: {precision}")

# Imprimir el reporte de clasificación
print("Reporte de Clasificación:")
print(classification_report(y_val, y_pred))
```
